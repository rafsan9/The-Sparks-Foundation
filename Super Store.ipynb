# Observation 1

```python
df['State'].value_counts()
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x='State',data=df ,order=df['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

# Observation 2

```python
df.hist(figsize=(8,8),bins=40)
plt.show()
```

# Observation 3

```python
fig,ax=plt.subplots(figsize=(10,8))
ax.scatter(df['Sales'],df['Profit'])
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
ax.set_title('Sales vs Profit')
plt.show()
```

```python
sns.lineplot(x='Discount',y='Profit',label= 'Profit',data=df)
plt.legend()
plt.show()
```

# Observation 4

```python
sns.lineplot(x='Quantity',y='Profit',label= 'Profit',data=df)
plt.legend()
plt.show()
```

```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['red','blue'],figsize=(8,6))
plt.ylabel('profit/loss and sales')
plt.show()
```

# Observation 5


```python
plt.figure(figsize=(12,10))
plt.title('Sales in Each Region')
sns.barplot(x='Region',y='Sales',data=df,hue='Segment',order=df['Region'].value_counts().index,palette='rocket')
plt.xlabel('Region',fontsize=10)
plt.show()
```

# Obsevation 6

```python
df.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['red','blue'],figsize=(8,6))
plt.ylabel('profit/loss and sales')
plt.show()
```

# Observation 7

```python
raf=df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
raf[:].plot.bar(color=['red','blue'],figsize=(8,6))
plt.title('profit/loss and sales across states')
plt.xlabel('States')
plt.ylabel('profit/loss and sales')
plt.show()
```

# Observation 8

```python
s_states=df['State'].value_counts().nlargest(10)

s_states
```

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['green','blue'],figsize=(8,6))
plt.ylabel('profit/loss and sales')
plt.show()
```

# Observation 9

```python
raf=df.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
raf[:].plot.bar(color=['red','blue'],figsize=(8,6))
plt.title('profit/loss and sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('profit/loss and sales')
plt.show()
```
